```python
import numpy as np
import pandas as pd
from sklearn.metrics.pairwise import euclidean_distances
from recomender.builders import DistanceBasedRewardBuilder 
```

```python
# Generate a synthetic dataset
np.random.seed(42)
data = {
    'numerical_feature': np.random.rand(10),
    'categorical_feature': ['cat', 'dog', 'fish', 'bird', 'lion', 'tiger', 'bear', 'wolf', 'fox', 'rabbit']
}
df = pd.DataFrame(data)
df.head()
```

```python
# Initialize the DistanceBasedRewardBuilder
reward_builder = DistanceBasedRewardBuilder(df, euclidean_distances)

# Generate the reward matrix
reward_matrix = reward_builder.build_reward_matrix()

# Display the reward matrix
print(reward_matrix)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Visualize the reward matrix
sns.heatmap(reward_matrix, annot=True, cmap='coolwarm')
plt.title('Reward Matrix Heatmap')
plt.xlabel('Item Index')
plt.ylabel('Item Index')
plt.show()
```
